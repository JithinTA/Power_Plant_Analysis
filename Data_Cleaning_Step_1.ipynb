# DATA CLEANING STEP 1

## The aim of this step is to fill any blank values in the global_power_plant_database.csv file and to convert it into a text format that can be then inserted into the SQL table.

## IMPORTING NECESSARY LIBRARIES

```python
import pandas as pd
import numpy as np
```

## IMPORTING DATA

```python
A = pd.read_csv("global_power_plant_database.csv", low_memory = False)
```

```python
A
```

## DATA CLEANING

### Deleting unnesessary columns

```python
cols = ["country", "latitude", "longitude", "other_fuel1", "other_fuel2", "other_fuel3", "owner"
       , "source", "url", "geolocation_source", "wepp_id", "year_of_capacity_data", "generation_gwh_2013",
          "generation_gwh_2014", "generation_gwh_2015", "generation_gwh_2016", "generation_gwh_2017",
          "generation_gwh_2018", "generation_gwh_2019", "generation_data_source", "estimated_generation_note_2013",
          "estimated_generation_note_2014", "estimated_generation_note_2015", "estimated_generation_note_2016",
          "estimated_generation_note_2017"]
A.drop(cols, inplace = True, axis = 1)
```

```python
A
```

### Filling blank spaces

```python
A.fillna(value = 0, inplace = True)
A
```

### Converting all the values in the dataframe to strings

```python
A = A.astype("string")
```

```python
A.dtypes
```

## CONVERTING THE DATAFRAME TO A TEXT FILE

### Converting the data frame to a list of tuples.

```python
Y = []
for i in range(len(A)):
    x = tuple(A.iloc[i])
    Y.append(x)  
    
```

```python
Y[9090]
```

### Writing the contents on to a text file

```python
file = open("power_plants1", "w", encoding = "utf-8")
for jock in Y:
    file.write(str(jock) + "," + "\n")
file.close()    
    
```
